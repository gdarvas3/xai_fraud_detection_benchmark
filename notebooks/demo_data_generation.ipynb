```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../data/raw/ieee_fraud.csv')
```

```python
df.shape
```

```python
CELVALTOZO_NEVE = 'isFraud' 
KERESETT_MINTA_MERET = 10000

kis_minta_df, _ = train_test_split(
    df,
    train_size=KERESETT_MINTA_MERET,
    stratify=df[CELVALTOZO_NEVE],
    random_state=42
)
```

```python
kis_minta_df
```

```python
kis_minta_df.to_csv('../data/raw/ieee_fraud_demo.csv', index=False)
```
