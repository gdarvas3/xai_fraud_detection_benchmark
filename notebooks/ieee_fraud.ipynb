# IEEE fraud dataset 

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/raw/ieee_fraud.csv')
```

```python
one_week_seconds = 7 * 24 * 3600

df_week = df[df["TransactionDT"] < one_week_seconds].copy()

df_week["Hour"] = df_week["TransactionDT"] // 3600

plt.figure(figsize=(10,5))
plt.hist(df_week["Hour"], bins=7*24, rwidth=0.8)
plt.xlabel("TransactionDT óránkénti bontásban")
plt.ylabel("Tranzakciók száma")
plt.title("Tranzakciók eloszlása az első héten óránként")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(df["TransactionAmt"], bins=100, log=True)
plt.xlabel("TransactionAmt")
plt.ylabel("Count (log)")
plt.title("Tranzakciók összegeinek eloszlása")
plt.show()
```
